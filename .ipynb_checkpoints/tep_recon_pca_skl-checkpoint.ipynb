```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score
from sklearn.manifold import TSNE
```

```python
df_trg_faultfree = pd.read_csv('./tep_data/tep_faultfree_trg.csv', index_col=0)
df_trg_faulty = pd.read_csv('./tep_data/tep_faulty_trg.csv', index_col=0)
df_trg_faultfree.shape, df_trg_faulty.shape
```

```python
df_trg = pd.concat([df_trg_faultfree, df_trg_faulty], axis=0)
df_trg.shape
```

```python
df_trg = df_trg[df_trg['simulationRun'] <= 5]
df_trg.shape
```

```python
scaler = StandardScaler()
X_faultfree = df_trg[df_trg['faultNumber']==0].iloc[:,3:]
X_faultfree_scaled = scaler.fit_transform(X_faultfree)
```

```python
exp_var = 0.90
```

```python
pca = PCA(n_components=exp_var, svd_solver='full')
X_faultfree_pca = pca.fit_transform(X_faultfree_scaled)
```

```python
print(pca.n_components_, np.sum(pca.explained_variance_ratio_))
```

```python
def recon_loss(pca,X):
    X_pca = pca.transform(X)
    X_recon = pca.inverse_transform(X_pca)
    loss = np.mean((X-X_recon)**2,axis=1)
    return loss
```

```python
loss_faultfree = recon_loss(pca,X_faultfree_scaled)
```

```python
loss_faultfree
```

```python
plt.figure(figsize=(6,3))
plt.hist(loss_faultfree, bins=100,label='Fault Free',alpha=1)
plt.title(f'Distribution of Reconstruction Error (PCA) Faultfree')
plt.xlabel('Reconstruction Error')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
loss_faulty = [loss_faultfree]
for id in range(1,21):
    df_fault=df_trg[(df_trg.faultNumber==id) & (df_trg['sample'] > 20 )]
    X_faulty = scaler.transform(df_fault.iloc[:,3:])
    loss = recon_loss(pca=pca,X=X_faulty)
    loss_faulty.append(loss)
```

```python
for i in range(1,21):
    plt.figure(figsize=(6,3))
    plt.hist(loss_faultfree, bins=100,label='Fault Free',alpha=1)
    plt.hist(loss_faulty[i], bins=100,label='Faulty',alpha=0.75)
    plt.title(f'Distribution of Reconstruction Error (PCA) Fault # {i}')
    plt.xlabel('Reconstruction Error')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

```python
fig, axes = plt.subplots(nrows=7, ncols=3, figsize=(15, 20))
for i, ax in enumerate(axes.flat):
    ax.hist(loss_faultfree, bins=100,label='Fault Free',alpha=1)
    if i > 0:
        ax.hist(loss_faulty[i], bins=100,label='Faulty',alpha=0.75)
    ax.set_title(f'Fault # {i}')
    ax.legend()
plt.tight_layout()
plt.show()
```

```python
X_pca = pca.transform(scaler.transform(df_trg.iloc[:,3:]))
```

```python
x_tsne = TSNE(n_components=2, learning_rate='auto',init='random', n_jobs=-1, verbose=2).fit_transform(X_pca[::5,:])
```

```python
y_label = df_trg.iloc[::5,0]
```

```python
f, ax = plt.subplots(figsize=(8,6))
sns.scatterplot(x=x_tsne[:,0],y=x_tsne[:,1],hue=y_label,style=y_label,palette=sns.color_palette(None, 21))
plt.legend(bbox_to_anchor=(1.15, 0.95))
plt.title('PCA')
plt.show()
```

```python
X_faultfree = df_trg[(df_trg['faultNumber']==0) & (df_trg['simulationRun'] == 1)].iloc[:,3:]
X_faultfree_scaled = scaler.fit_transform(X_faultfree)
loss_faultfree = recon_loss(pca,X_faultfree_scaled)
loss_faulty = [loss_faultfree]
for id in range(1,21):
    df_fault=df_trg[(df_trg.faultNumber==id) & (df_trg['simulationRun'] == 1)]
    X_faulty = scaler.transform(df_fault.iloc[:,3:])
    loss = recon_loss(pca=pca,X=X_faulty)
    loss_faulty.append(loss)
```

```python
mu = np.mean(loss_faultfree)
sigma = np.std(loss_faultfree)
threshold = mu+3*sigma
threshold
```

```python
fig, axes = plt.subplots(nrows=7, ncols=3, figsize=(15, 20))
for i, ax in enumerate(axes.flat):
    ax.plot(loss_faulty[i])
    ax.set_title(f'Fault # {i}')
    ax.axhline(threshold,c='r',label='threshold')    #threshold value
    ax.axvline(20,c='m') #time of fault introduction
    ax.legend()
plt.tight_layout()
plt.show()
```

```python
print(f'Variance {exp_var}')
print(pca.n_components_, np.sum(pca.explained_variance_ratio_))
list_accuracy = []
list_f1 = []
list_recall = []
list_precision = []
for id in range(1,len(loss_faulty)):
    y_pred = (loss_faulty[id] > threshold).astype(int)
    if id==0:
        y_true = np.repeat(0, len(y_pred))
    else:
        y_true = np.repeat(1, len(y_pred))
        y_true[0:21]=0
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred, average='macro')
    precision = precision_score(y_true, y_pred, average='macro')
    print(f'{id}\taccuracy\t{accuracy:.2%}\tf1_score\t{f1:.2%}\trecall\t{recall:.2%}\tprecision\t{precision:.2%}')
    list_accuracy.append(accuracy)
    list_f1.append(f1)
    list_recall.append(recall)
    list_precision.append(precision)
print(f'accuracy\t: {np.mean(np.array(list_accuracy)):.2%}')
print(f'f1_score\t: {np.mean(np.array(list_f1)):.2%}')
print(f'recall  \t: {np.mean(np.array(list_recall)):.2%}')
print(f'precision\t: {np.mean(np.array(list_precision)):.2%}')
```
